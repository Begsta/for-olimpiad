```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from keras_visualizer import visualizer
import pandas as pd
import numpy as np
from tqdm.keras import TqdmCallback
import graphviz
from ta.volume import ChaikinMoneyFlowIndicator as cmf,ForceIndexIndicator as fi, MFIIndicator as MFI
from ta.momentum import RSIIndicator as rsi, StochRSIIndicator as SRSI
from ta.trend import MassIndex as MS

def normalisation(column):
    column = (column - min(column))
    column /= max(column)
    return column



df = pd.read_csv('SBER_190218_230218.csv', sep=';')
df = df[['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>', '<VOL>']]
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

W = 14
df['cmf'] = cmf(high=df['High'], low=df['Low'], volume=df['Volume'], close=df['Close'] , fillna=False, window=W).chaikin_money_flow()
df['fi'] = fi(volume=df['Volume'], close=df['Close'] , fillna=False, window=W).force_index()
df['rsi'] = rsi(close=df['Close'] , fillna=False, window=W).rsi()
df["MFI"] = MFI(high=df['High'], low=df['Low'], volume=df['Volume'], close=df['Close'] , fillna=False, window=W).money_flow_index() 
df['SRSI'] = SRSI(close=df['Close'], fillna=False, window=W).stochrsi()
df['MS'] = MS(high=df["High"], low=df['Low'], fillna=False).mass_index()
A = np.array(df['Close'][1:]) - np.array(df['Close'][:-1])
A = list(A) + [0]



df['price_change'] = A
df = df.dropna().reset_index().drop(columns=['index'])[:-1]
for i in df.columns[:-1]:
    df[i] = normalisation(df[i])
array_x = np.array(df[["cmf",'rsi', 'MFI', 'SRSI', 'MS']])

array_up, array_low = [0]*len(df['price_change']), [0]*len(df['price_change'])
for i in range(len(df['price_change'])-1):
    if df['price_change'][i+1] > 0:
        array_up[i] = 1; array_low[i] = 0
    else:
        array_up[i] = 0; array_low[i] = 1
df['up'] = array_up
df['low'] = array_low
df=df[20:-2]
array_y = df[[ 'up', 'low']]
df
result[0] > 0.6
```

```python
neural_network = Sequential()

neural_network.add(  Dense(50, input_dim=len(df.iloc[0])-9, activation='linear')  )
neural_network.add(  Dense(20, activation='relu')  )
neural_network.add( Dense(20, activation= 'relu') )
neural_network.add( Dense(20, activation= 'relu') )
neural_network.add(  Dense(2, activation='softmax')  )

#neural_network.summary()
neural_network.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tqdm.keras import TqdmCallback
N = len(df)
train = int(9/10*N)
test = N-train

train_data, train_res = array_x[:train], array_y[:train]
test_data, test_res = array_x[-test:], array_y[-test:]
neural_network.fit(train_data, train_res #тренируем нейронную сеть
          , epochs=100 #количество повторов во время обучения
          , validation_split = 0.05
          , batch_size=64
          , verbose=0, callbacks=[TqdmCallback(verbose=0)]) #красивый вывод обучения
```

```python
neural_network.evaluate(test_data, test_res) #проверка точности
```

```python
#Импорт библиотек
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.linear_model import LinearRegression
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
from sklearn.model_selection import GridSearchCV
import numpy as np
import warnings
from tqdm import tqdm
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from keras_visualizer import visualizer
from tqdm.keras import TqdmCallback
import keras
warnings.filterwarnings("ignore")

#функция для нормализации колон
def normalisation(column):
    column = (column - min(column))
    column /= max(column)
    return column

#Считывание файла train.scv
df_titanic_train = pd.read_csv('train.csv')

#распределение колонок Pclass
oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(df_titanic_train[["Pclass"]])
df_oe_results = pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_)
df_titanic_train['Pclass1'] = df_oe_results[1]
df_titanic_train['Pclass2'] = df_oe_results[2]
df_titanic_train['Pclass3'] = df_oe_results[3]

#Убираем всё лишнее
df_titanic_train = df_titanic_train[['Sex', 'Age', 'Pclass1', 'Pclass2', 'Pclass3', 'Survived']]

#Перемешиваем столбцы для более хорошего результата
df_titanic_train = df_titanic_train.sample(frac=1)

#переделываем столбец пола в 0 для мужчин, а 1 для женщин
for i in range(len(df_titanic_train['Sex'])):
    if df_titanic_train['Sex'][i] == 'male':
        df_titanic_train["Sex"][i] = 0
    else:
        df_titanic_train["Sex"][i] = 1

#убираем все столбцы в которых есть значение NaN
df_titanic_train.dropna(inplace=True)

#Нормализуем столбцы 
for i in df_titanic_train.columns[:-1]:
    df_titanic_train[i] = normalisation(df_titanic_train[i])

#дулим данные на тренеровочные и тестовые
x = df_titanic_train[[ 'Sex', 'Age', 'Pclass1', 'Pclass2', 'Pclass3']]
y = df_titanic_train ['Survived']
x_titanic_train, x_titanic_test = x[:600], x[600:]
y_titanic_train, y_titanic_test = y[:600], y[600:]

#создаём и тестируем модель логической регресии 
model1 = LogisticRegression()
model1.fit(x_titanic_train, y_titanic_train)
y_pred = model1.predict(x_titanic_test)
print('Score логической регрессии: ', model1.score(x_titanic_test, y_titanic_test))

#создание регрессии с множеством точек
model = KNeighborsClassifier(2)
model.fit(x_titanic_train, y_titanic_train)
y_pred = model.predict(x_titanic_test)
print('Score регрессии с множеством точек: ', model.score(x_titanic_test, y_titanic_test))

#поиск лучшего результата в двух моделях, для понимания какая модель лучше
scores = cross_val_score(model1, x, y, cv=5)
print("Средний резултат в моделе с линейной регрессией: ", np.mean(scores))
scores = cross_val_score(model, x, y, cv=5)
print("Средний результат в моделе с N множеством точек: ", np.mean(scores))

#создание cvs модели
clf = svm.SVC(kernel='poly', degree=4)

#проверка на обучение
scores = cross_val_score(clf, x, y, cv=5)
print('Средний результат в моделе svc: ', np.mean(scores))

#Параметры для каждой модели
parameters_cvs = {
    'kernel':('linear', 'poly', 'rbf', 'sigmoid'),
              'C':[i for i in range(1, 16)],
              'degree':[i for i in range(1, 11)],
             'gamma': ['scale', 'auto']
             }
parametrs_LR ={'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'penalty':['none', 'elasticnet', 'l1', 'l2'],
              'C':[0.001, 0.01, 0.1, 1, 10, 100]}
parametrs_KNC = {
        'n_neighbors':[2],
        'weights' : ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute' ]
    }
#Пересоздаём модель 
x_titanic_train = x_titanic_train.astype(float)
y_titanic_train = y_titanic_train.astype(float)
x_titanic_test = x_titanic_test.astype(float)
y_titanic_test = y_titanic_test.astype(float)
```

```python
model_LR = LogisticRegression()
model_KNN = KNeighborsClassifier()
model_cvs = svm.SVC()
def GridSearch(models, x, y):
    good_score = {}
    for i, j in tqdm(models.items()):
        clf = GridSearchCV(i, j)
        clf.fit(x, y)
        good_score[i] = clf.best_score_
    print(good_score)
GridSearch({
        model_LR : parametrs_LR,
    model_KNN : parametrs_KNC,
    model_cvs : parameters_cvs
    },x, y
)
```
